```python
#Ask the user to enter distance in kilometers
km = float(input("Enter value in km:"))

#Define the conversion factor (1 km ≈ 0.621371 miles)
conversion_factor = 0.621

#Multiply kilometers by the conversion factor to get miles
m = km * conversion_factor

#Display the result
print(f"{km} kilometers is equal to {m} miles")
```

```python
#Calculate the area of a circle

import math

#Define the radius
radius = 10

#Calculate the area of the circle using this formula: π×r2
area = math.pi * (radius ** 2)

print(f"The area of the circle with radius", radius, "is" , area)
```

```python
#Check if a number is even or odd
number = int(input("Enter number:"))

#checking if number is even or odd using conditional statements
if number % 2 == 0:
    print(number, "is even")
else:
    print(number, "is odd")



```
