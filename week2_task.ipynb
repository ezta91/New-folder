```python
num1 = float(input("Enter the first number:"))

operation = input("Enter an operation (+, -, *, /):")

num2 = float(input("Enter the second number:"))

if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    if num2 != "0":
        result = num1 / num2
    else:
        result = "Error! Division by 0"
else:
    result = "Invalid Operation!"

print("Result:", result)
```

```python
#Ask the user to enter distance in kilometers
km = float(input("Enter value in km:"))

#Define the conversion factor (1 km ≈ 0.621371 miles)
conversion_factor = 0.621

#Multiply kilometers by the conversion factor to get miles
m = km * conversion_factor

#Display the result
print(f"{km} kilometers is equal to {m} miles")
```

```python
#Ask the user to enter weight in pounds
ib = float(input("Enter value in pounds:"))

##Define the conversion factor (1 Ib ≈ 0.453592 kg)
conversion_factor1 = 0.454

#Multiply pounds by the conversion factor to get kilograms
kg = ib * conversion_factor1

#Display result
print(f"{ib} pounds is equal to {kg} kilogram")
```

```python
#Ask the user to enter temperature in Celsius
C = float(input("Enter value in Celsius:"))

#Define the conversion factor
F = (C * 1.8) + 32

#Display result
print(f"{C} Celsius is equal to {F} Fahrenheit")
```

```python
#Ask user to enter energy in Joules
J = float(input("Enter energy in Joules"))

#Define the conversion factor (0.000239006)
kcal = J * 0.000239006

#Display result
print(f"{J} Joules is equal to {kcal} kilocalorie")
```
